```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_movies.csv')
genres=pd.read_csv('tmdb_genres.csv')
movies
```

```python
genres
```

```python
#Below Im merging column genre_id with df movies
```

```python
df = movies.merge(genres,left_on='genre_id', right_on='Unnamed: 0')
df                                                       
```

```python
#Im selecting movies of which vote_count is higher than 3rd quantile of the vote distribution
```

```python
df75 = df[df['vote_count'] > df['vote_count'].quantile(.75)]
df75
```

```python
#Now Im returning a list of the top 10 vote_average with vote_count greater than the 3rd quartile of the vote distribution.
```

```python
df75['vote_average'].nlargest(n=10)
```

```python
# I group the table to get the average revenue and average budget per year for videos published from 2010 to 2016
```

```python

df3 = df.loc[df["release_date"].between("2010-01-01", "2016-12-31")]
df3
```

```python
df4 = df3.groupby("Year", as_index=True)[['revenue', 'budget']].mean().reset_index()
df4
```

```python
#Probowałam tutaj zastosować df4['revenue'].round(2) ale nie zadziałało, nie jestem pewna dlaczego
```

```python
df4['revenue'].round(2)
```

```python
import matplotlib.pyplot as plt

y1 = df4.revenue.plot(kind= 'bar',  y='revenue')
y1 = df4.budget.plot(kind= 'line', color='red',  y='budget')
plt.legend(bbox_to_anchor=(1.3, 1), loc='upper right', borderaxespad=0)

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Średni przychód i budżet filmów w latach 2010-2016")



```
