```python
df11=df[['RecruitmentSource','DateofHire','DateofTermination']]
df11
```

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')
```

```python
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format="%d/%m/%Y")
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='mixed')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df12=df[['DateofHire','DateofTermination','Seniority']]
df12
```

```python
conc=pd.concat([df11, df12], axis=1)
conc
```

```python
df13=conc[['RecruitmentSource','Seniority',]]
df13
```

```python
df14=df13.groupby('RecruitmentSource',as_index=True).mean()
df14

```
